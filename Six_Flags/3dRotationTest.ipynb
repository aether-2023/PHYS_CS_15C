```python
def calibrator(target_list, t_list):
    index = 1500 # use first 1500 measures to calibrate
    calibration_a = sum(target_list[:index])/index
    calibrated_list = target_list - calibration_a
    return calibrated_list
```

```python
def get_1dtraj(accel_list, t_interval_list, threshold=0.004):
    v_list = [0]
    for index in range(len(accel_list)):
        if index>0:
            v = t_interval_list[index]*accel_list[index]+v_list[index-1]
            if abs(v)<threshold:
                v_list.append(0)
            else:
                v_list.append(v)
    v_list = np.array(v_list)
    x_list = [0]
    for index in range(len(accel_list)):
        if index>0:
            x = (t_interval_list[index]**2)*accel_list[index]/2+v_list[index]*t_interval_list[index]+x_list[index-1]
            x_list.append(x)
    x_list = np.array(x_list)
    return v_list, x_list
```

```python
accel_x_abs_list = calibrator(accel_x_abs_list, t_list)
accel_y_abs_list = calibrator(accel_y_abs_list, t_list)
accel_z_abs_list = calibrator(accel_z_abs_list, t_list)
```

```python
vx_list, x_list = get_1dtraj(accel_x_abs_list, t_interval_list)
vy_list, y_list = get_1dtraj(accel_y_abs_list, t_interval_list)
vz_list, z_list = get_1dtraj(accel_z_abs_list, t_interval_list)
```
