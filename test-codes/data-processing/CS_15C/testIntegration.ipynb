```python
import numpy as np
import os

#initialization 
acceX = []
acceY = []
veloX = []
veloY = []
posiX = []
posiY = []
acceX.append(0)
acceY.append(0)
veloX.append(0)
veloY.append(0)
posiX.append(0)
posiY.append(0)
intervalT = np.linspace(0,10,10001)
delT = 0.0001

#take data
for i in range(10000):
    acceX.append(np.cos(i*2*np.pi/10000))
    acceY.append(np.sin(i*2*np.pi/10000))
    #(-0.5+int.from_bytes(os.urandom(8), byteorder="big") / ((1 << 64) - 1))

#integration
for i in range(10000):
    veloX.append(veloX[i] + acceX[i] * delT)
    veloY.append(veloY[i] + acceY[i] * delT)

for i in range(10000):
    posiX.append(posiX[i] + veloX[i] * delT) #+ 0.5*acceX[i] * (delT**2))
    posiY.append(posiY[i] + veloY[i] * delT) #+ 0.5*acceY[i] * (delT**2))

```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.plot(intervalT, acceY, label = 'acceY')
plt.plot(intervalT, veloY, label = 'veloY')
plt.plot(intervalT, posiY, label = 'posiY')
plt.xlabel('t')
plt.legend()
plt.show()
```

```python
plt.figure()
plt.plot(intervalT, acceX, label = 'acceX')
plt.plot(intervalT, veloX, label = 'veloX')
plt.plot(intervalT, posiX, label = 'posiX')
plt.xlabel('t')
plt.legend()
plt.show()
```

```python
plt.figure()
plt.plot(posiX, posiY, label = 'path')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
